```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure plots are displayed inline
%matplotlib inline

# Step 1: Define a list of stock ticker symbols for multiple companies
tickers = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'META', 'TSLA', 'NFLX', 'NVDA', 'BRK-B', 'JPM']

# Initialize a dictionary to store data for each company
company_data = {}

# Loop through each ticker
for ticker in tickers:
    print(f"\nProcessing {ticker}...")
    
    # Step 2: Fetch historical stock price data
    data = yf.download(ticker, start='2010-01-01', end='2023-01-01')

    # Check if data is fetched
    if data.empty:
        print(f"No data found for {ticker}.")
        continue

    print(f"Data shape for {ticker}: {data.shape}")
    print(data.head())  # Display the first few rows of the data

    # Step 3: Get company name
    stock_info = yf.Ticker(ticker)
    company_name = stock_info.info.get('longName', 'Unknown Company')  # Fetch the full company name

    # Display the company name
    print(f"Company Name: {company_name}")

    # Store the data in the dictionary
    company_data[ticker] = (company_name, data)

    # Step 4: Data Preprocessing
    data['SMA_5'] = data['Close'].rolling(window=5).mean()
    data['SMA_20'] = data['Close'].rolling(window=20).mean()
    data['Return'] = data['Close'].pct_change()
    data.dropna(inplace=True)

    # Step 5: Exploratory Data Analysis (EDA)
    plt.figure(figsize=(14, 7))
    plt.plot(data['Close'], label='Close Price', color='blue')
    plt.plot(data['SMA_5'], label='5-Day SMA', color='orange')
    plt.plot(data['SMA_20'], label='20-Day SMA', color='green')
    plt.title(f'{company_name} Stock Price with Moving Averages')
    plt.legend()
    plt.show()

    # Correlation Analysis
    numeric_data = data.select_dtypes(include=[np.number])
    correlation_matrix = numeric_data.corr()
    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
    plt.title(f'Correlation Matrix for {company_name}')
    plt.show()
```
