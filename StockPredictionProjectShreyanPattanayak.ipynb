```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure plots are displayed inline
%matplotlib inline

# Step 1: Define a list of stock ticker symbols for multiple companies
tickers = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'META', 'TSLA', 'NFLX', 'NVDA', 'BRK-B', 'JPM']

# Initialize a dictionary to store data for each company
company_data = {}

# Loop through each ticker
for ticker in tickers:
    print(f"\nProcessing {ticker}...")
    
    # Step 2: Fetch historical stock price data
    data = yf.download(ticker, start='2010-01-01', end='2023-01-01')

    # Check if data is fetched
    if data.empty:
        print(f"No data found for {ticker}.")
        continue

    print(f"Data shape for {ticker}: {data.shape}")
    print(data.head())  # Display the first few rows of the data

    # Step 3: Get company name
    stock_info = yf.Ticker(ticker)
    company_name = stock_info.info.get('longName', 'Unknown Company')  # Fetch the full company name

    # Display the company name
    print(f"Company Name: {company_name}")

    # Store the data in the dictionary
    company_data[ticker] = (company_name, data)

    # Step 4: Data Preprocessing
    data['SMA_5'] = data['Close'].rolling(window=5).mean()
    data['SMA_20'] = data['Close'].rolling(window=20).mean()
    data['Return'] = data['Close'].pct_change()
    data.dropna(inplace=True)

    # Step 5: Exploratory Data Analysis (EDA)
    plt.figure(figsize=(14, 7))
    plt.plot(data['Close'], label='Close Price', color='blue')
    plt.plot(data['SMA_5'], label='5-Day SMA', color='orange')
    plt.plot(data['SMA_20'], label='20-Day SMA', color='green')
    plt.title(f'{company_name} Stock Price with Moving Averages')
    plt.legend()
    plt.show()

    # Correlation Analysis
    numeric_data = data.select_dtypes(include=[np.number])
    correlation_matrix = numeric_data.corr()
    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
    plt.title(f'Correlation Matrix for {company_name}')
    plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dropout, Dense
import numpy as np
import matplotlib.pyplot as plt

# Assuming 'data' is already defined and contains the necessary columns

# Step 6: Prepare Data for Random Forest
X = data[['SMA_5', 'SMA_20', 'Return']]
y = data['Close'].values.flatten()  # Ensure y is a 1D array

# Split the data into training and testing sets
X_train, X_test = X[:-100], X[-100:]
y_train, y_test = y[:-100], y[-100:]

# Step 7: Train Random Forest Model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Step 8: Make Predictions with Random Forest
rf_predictions = rf_model.predict(X_test)

# Step 9: Evaluate Random Forest Model
rf_mse = mean_squared_error(y_test, rf_predictions)
print(f'Random Forest Mean Squared Error: {rf_mse}')

# Step 10: Prepare Data for LSTM
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data['Close'].values.reshape(-1, 1))

# Split the scaled data into training and testing sets
train_size = int(len(scaled_data) * 0.8)
train_data = scaled_data[:train_size]
test_data = scaled_data[train_size:]

# Function to create dataset for LSTM
def create_dataset(data, time_step=1):
    X, y = [], []
    for i in range(len(data) - time_step - 1):
        X.append(data[i:(i + time_step), 0])
        y.append(data[i + time_step, 0])
    return np.array(X), np.array(y)

# Create datasets for LSTM
time_step = 60
X_train_lstm, y_train_lstm = create_dataset(train_data, time_step)
X_test_lstm, y_test_lstm = create_dataset(test_data, time_step)

# Reshape input to be [samples, time steps, features]
X_train_lstm = X_train_lstm.reshape(X_train_lstm.shape[0], X_train_lstm.shape[1], 1)
X_test_lstm = X_test_lstm.reshape(X_test_lstm.shape[0], X_test_lstm.shape[1], 1)

# Step 11: Build and Train LSTM Model
lstm_model = Sequential()
lstm_model.add(LSTM(50, return_sequences=True, input_shape=(X_train_lstm.shape[1], 1)))
lstm_model.add(Dropout(0.2))
lstm_model.add(LSTM(50, return_sequences=False))
lstm_model.add(Dropout(0.2))
lstm_model.add(Dense(1))

# Compile the model
lstm_model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
lstm_model.fit(X_train_lstm, y_train_lstm, epochs=50, batch_size=32)

# Step 12: Make Predictions with LSTM
lstm_predictions = lstm_model.predict(X_test_lstm)
lstm_predictions = scaler.inverse_transform(lstm_predictions)  # Inverse scaling

# Step 13: Evaluate LSTM Model
# Reshape y_test_lstm to match the shape of lstm_predictions
y_test_lstm = y_test_lstm.reshape(-1, 1)  # Ensure y_test_lstm is 2D for inverse scaling
y_test_lstm = scaler.inverse_transform(y_test_lstm)  # Inverse scaling

lstm_mse = mean_squared_error(y_test_lstm, lstm_predictions)
print(f'LSTM Mean Squared Error: {lstm_mse}')

# Step 14: Visualize Predictions
plt.figure(figsize=(14, 7))
plt.plot(data.index[-len(y_test):], y_test, label='Actual Price', color='blue')
plt.plot(data.index[-len(lstm_predictions):], lstm_predictions, label='LSTM Predictions', color='red')
plt.title(f'{company_name} Stock Price Prediction')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```
